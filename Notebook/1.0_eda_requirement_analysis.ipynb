```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os

# Use '..' to go up to the root, then into 'data'
file_path = '../data/WA_Fn-UseC_-Telco-Customer-Churn.csv'

df = pd.read_csv(file_path)

# The "Trap": Converting TotalCharges to numeric
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')

# Check for the nulls we just created
print(f"Missing values in TotalCharges: {df['TotalCharges'].isnull().sum()}")
df.dropna(inplace=True) # Drop the few rows with missing TotalCharges
```

```python
plt.figure(figsize=(6, 4))
sns.countplot(x='Churn', data=df, palette='viridis')
plt.title('Class Distribution (Churn vs No Churn)')
plt.show()

print(df['Churn'].value_counts(normalize=True))
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='Contract', hue='Churn', data=df)
plt.title('Churn Rate by Contract Type')
plt.show()
```

```python
sns.boxplot(x='Churn', y='tenure', data=df)
```
